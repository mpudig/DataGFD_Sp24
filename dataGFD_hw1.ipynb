My class project for the data-driven GFD class (Spring 2024) will focus on marine heatwave (MHW) predictability and forecasting, with a focus on the effects of ENSO (e.g., its effect on changing predictability and forecast skill).

---

MHWs are defined as a prologned, discrete, anomalously warm water events (usually anomalously warm SST events) in a particular location. Specifically, a MHW is said to occur if it persists for at least five days, has a well-defined start and end point, and is warmer than the 90th percentile of a 30-year climatology (Hobday et al., 2016). They can have profound impacts both ecologically and socioeconomically. MHWs can cause declines in marine primary productivity (i.e., photosynthesis), coral bleaching events, mass die-offs, and large geographical shifts in species habitats; they can also lead to harmful algal blooms and early peaks in species populations, which can disrupt the operations of fishery and other ocean-based industries. However, understanding of MHW drivers as well as their predictability is still a nascent research field. Given their potential for significant ecological and economic disruption, improving our ability to anticipate such events is crucial.

---

A recent paper analyzed data from a set of large-ensemble climate model simulations and analyzed the occurrences of MHWs (as well as marine coldwaves) therein (Deser et al., 2024). These simulations include 30 ensemble members (branched from different points of the _piControl_) under historical and RCP8.5 forcing. ### Calculate internal SST components

In order to separate changes in MHWs as arising from (i) changes in variability and (ii) changes in the mean state, Deser et al. (2024) define the internal component of SST for each model as the ensemble-mean anomalized SST for each model, i.e., $$\text{SST}_{i}(x, y, t; e) = \text{SST}(x, y, t; e) - \text{SST}(x, y, t; em),$$ where $\text{SST}(x, y, t; e)$ is the SST of ensemble member $e$ and $\text{SST}(x, y, t; em)$ is the ensemble-mean.

```python
'''
To create ensemble mean, add all tos dataarrays to a dataset and then take mean over ensemble members?
'''

tos_ensemble = xr.zeros_like(tos_dict[paths[0]]).tos
tos_ensemble = tos_ensemble.expand_dims(dim = 'ensemble', axis = 0)
tos_ensemble_list = [tos_dict[path].tos for path in paths]
tos_ensemble = xr.concat(tos_ensemble_list, dim = 'ensemble')
tos_em = tos_ensemble.mean('ensemble')
```

```python
def internal_SST(da, da_em):
    
    return da - da_em
```

### Calculate ENSO-related SST pattern in each model

To calculate the ENSO-related pattern in each model, I again follow the procedure of Deser et al. (2024). I compute the leading EOF (EOF1) of 3-month running means of $\text{SST}_i$ for each model over a tropical Pacific domain (10$^\circ$N-10$^\circ$S, 120$^\circ$E-85$^\circ$W) for three 30-year periods (1970-2000, 2020-2050, 2070-2100). 

```python
def rolling_mean(da, period):

    return da.rolling(time = period, center=True).mean().isel(time = slice(3, -3))

def annual_av(da):

    return da.resample(time = 'A').mean(dim = 'time')
    
def ENSO_region(da):

    return da.sel(lat = slice(-5, 5), lon = slice(170, 360 - 120))
```

```python
### Do EOF-PC analysis on the annually-averaged, ENSO-regioned fields

# Field
tos_1970_2000_ENSO = ENSO_region(rolling_mean(internal_SST(tos_dict[paths[0]].tos.sel(time = slice("1970", "2000")), tos_em.sel(time = slice("1970", "2000"))), 5))
# tos_1970_2000_ENSO = ENSO_region(annual_av(internal_SST(tos_dict[paths[0]].tos.sel(time = slice("1970", "2000")), tos_em.sel(time = slice("1970", "2000")))))
# tos_1970_2000_ENSO = ENSO_region(tos_dict[paths[0]].tos.sel(time = slice("1970", "2000")))
# tos_1970_2000_ENSO = annual_av(tos_1970_2000_ENSO) - tos_1970_2000_ENSO.mean('time')
tos_1970_2000_ENSO = tos_1970_2000_ENSO * np.sqrt(np.cos(tos_1970_2000_ENSO.lat * np.pi / 180))

# Get shape
N, Y, X = tos_1970_2000_ENSO.shape

# Flatten into time x (lon * lat)
tos_1970_2000_ENSO_flattened = tos_1970_2000_ENSO.values.reshape(N, Y * X) # Each column is a month

# Compute the spatial covariance matrix
cov_sst = np.cov(tos_1970_2000_ENSO_flattened, rowvar = False)

# Compute the eigenvalues and eigenvectors
lam, E = la.eig(cov_sst)
lam, E = np.real(lam), np.real(E)

def PCA_order(order):
    e = E[:, order - 1]
    u = (e * tos_1970_2000_ENSO_flattened).sum(axis = 1)

    # Multiply, reshape, and make into an xarray dataarray
    # Probably a better way to do this last "full_like" bit with xarray, but I cannot remember easy syntax...

    PCA_reconstruct = np.outer(u, e).reshape(N, Y, X) * xr.full_like(tos_1970_2000_ENSO, 1)

    # Divide by cosine of latitude to remove weighting used earlier
    PCA_reconstruct = PCA_reconstruct / np.sqrt(np.cos(tos_1970_2000_ENSO.lat * np.pi / 180))

    return PCA_reconstruct


def EOF(order):
    return E[:, order]
```

```python
plt.figure(figsize=(10,6))
fontsize = 16
plt.plot(np.arange(1,11), np.cumsum(lam[0:10]) / (np.sum(lam)))
plt.title('Proportion of variation accounted for EOFs', fontsize = fontsize + 2)
plt.ylabel('$R^2$', fontsize = fontsize)
plt.xlabel('Number of EOFs used', fontsize = fontsize)
plt.xticks(fontsize = fontsize)
plt.yticks(fontsize = fontsize)
plt.grid()
```

```python
EOF1, EOF2, EOF3, EOF4, EOF5, EOF6, EOF7 = PCA_order(1), PCA_order(2), PCA_order(3), PCA_order(4), PCA_order(5), PCA_order(6), PCA_order(7)
EOF7_reconstruct = EOF1 + EOF2 + EOF3 + EOF4 + EOF5 + EOF6 + EOF7
```
